```python
import pandas as pd
dataset = pd.read_csv(r"C:\Users\lenovo\OneDrive\Desktop\università\Magistrale\Primo Anno\Primo Semestre\Data Mining\train.csv")
dataset.head()
```

```python
# main stats of dataset
dataset.describe()
```

```python
# main stats of duration_ms and popularity
dataset[['duration_ms', 'popularity']].describe()
```

```python
# frequency distributions of duration_ms and popularity
dataset[['duration_ms', 'popularity']].hist()
```

```python
# main stats of genres
genres_dataset = dataset.groupby('genre')
genres_dataset.mean()
```

```python
# number of songs for each genre
dataset['genre'].value_counts().plot(kind='bar')
```

```python
# pie chart of "explicit"

import matplotlib.pyplot as plt

# count the number of occurrences of each value in the "explicit" column
explicit_counts = dataset['explicit'].value_counts()

# create a pie chart of the "explicit" column
explicit_counts.plot(kind='pie', autopct='%1.1f%%')

# add a title to the chart
plt.title('Explicit')
```

```python
# describe "duration_ms" and "popularity" for explicit = True
dataset[dataset['explicit'] == True][['duration_ms', 'popularity']].describe()
```

```python
# describe "duration_ms" and "popularity" for explicit = False
dataset[dataset['explicit'] == False][['duration_ms','popularity']].describe()
```

```python
# how many songs each artist has
artists_counts = dataset['artists'].value_counts()
print(artists_counts[artists_counts > 1])
```

```python
# bar chart of the number of songs for each artist

dataset['artists'].value_counts().plot(kind='bar')
plt.xlabel('Artists')
plt.ylabel('Number of Songs')
plt.show()
```

```python
# how many different artists are there in the dataset
dataset['artists'].nunique()
```

```python
# genre with the highest average popularity
popular_genre = dataset.groupby('genre').mean().sort_values(['popularity'], ascending=False).iloc[0].name
print(popular_genre)
```

```python
# genre with the highest average duration_ms
duration_genre = dataset.groupby('genre').mean().sort_values(['duration_ms'], ascending=False).iloc[0].name
print(duration_genre)
```

```python
# track more popular, longer and shorter
print(dataset.sort_values('popularity', ascending=False).iloc[0]['name'])
print(dataset.sort_values('duration_ms', ascending=False).iloc[0]['name'])
print(dataset.sort_values('duration_ms').iloc[0]['name'])
```

```python
# frequency distribution of "duration_ms" for explicit = True and explicit = False

import matplotlib.pyplot as plt

# histogram of "duration_ms" for Explicit=True
dataset[dataset['explicit'] == True]['duration_ms'].hist()
plt.title('Explicit=True')
plt.xlabel('Duration (ms)')
plt.ylabel('Count')
plt.show()

# histogram of "duration_ms" for Explicit=False
dataset[dataset['explicit'] == False]['duration_ms'].hist()
plt.title('Explicit=False')
plt.xlabel('Duration (ms)')
plt.ylabel('Count')
plt.show()
```

```python
# frequency distribution of "popularity" for explicit = True and explicit = False

import matplotlib.pyplot as plt

# histogram of "popularity" for Explicit=True
dataset[dataset['explicit']==True]['popularity'].hist()
plt.title('Explicit=True')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()

# histogram of "popularity" for Explicit=False
dataset[dataset['explicit']==False]['popularity'].hist()
plt.title('Explicit=False')
plt.xlabel('Popularity')
plt.ylabel('Count')
plt.show()
```
